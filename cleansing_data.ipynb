```python
import matplotlib.pyplot as plt
```

```python
plt.title('Top 10 Most Happiness Country in 2019')
df = df2019.head(10)
plt.plot(df['country'], df['happiness_score'], label='Happiness Score')
plt.bar(df['country'], df['freedom'], label='Freedom')
plt.xlabel("Country")
plt.ylabel("Happiness Score")
plt.show()
```
